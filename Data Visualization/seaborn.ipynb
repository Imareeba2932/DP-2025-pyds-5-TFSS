```python
import seaborn as sns
```

```python
df = sns.load_dataset('titanic')
df
```

```python
#Age Distribution
fig = sns.histplot(df, x='age')
fig.set_title("Age Distribution of Titanic Passenger")
fig.set_xlabel("Age")
fig.set_ylabel("frequency")
```

```python
# survival count based on gender
fig = sns.countplot(df, x='survived',hue='sex')
fig.set_title("Survival Count based on gender")
fig.set_xlabel("Survived")
fig.set_ylabel("Count")
```

```python
fig = sns.barplot(df, x='sex',y='fare', hue='class')
fig.set_title('Average Fare by Passenger Class and Gender')
fig.set_xlabel('Sex')
fig.set_ylabel('Average Fare')
```

```python
fig = sns.countplot(df, x='class', palette='coolwarm')
fig.set_title('Count of Passenger by Class')
fig.set_xlabel("Passenger Class")
fig.set_ylabel("Count")
```

```python
sns.pairplot(df[['age','fare','survived','pclass']],hue='survived',palette='terrain')
```
